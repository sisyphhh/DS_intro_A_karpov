```python
#stud_perform=pd.read_csv('C:\\Users\\poll\\zzz_Data_science\\StudentsPerformance.csv')
data=data = pd.DataFrame({'X_1': [1, 1, 1, 0, 0, 0, 0, 1], 
                          'X_2': [0, 0, 0, 1, 0, 0, 0, 1], 
                          'Y': [1, 1, 1, 1, 0, 0, 0, 0]})
data
```

```python
clf=tree.DecisionTreeClassifier(criterion='entropy')
```

```python
clf
```

```python
X = data[['X_1','X_2']]
y = data.Y
```

```python
clf.fit(X,y)
```

```python
import sklearn
print('The scikit-learn version is {}.'.format(sklearn.__version__))

tree.plot_tree(clf , feature_names=list(X),class_names=['Negative','Positive'], filled=True)
#tree.plot_tree(clf)
```
