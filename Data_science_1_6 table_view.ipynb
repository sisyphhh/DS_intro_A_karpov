```python
stud_perform=pd.read_csv(directory_1+'PD_pract\\StudentsPerformance.csv')
```

```python
stud_perform = stud_perform \
    .rename(columns = 
               {'parental level of education':'parental_level_of_education',
                'test preparation course':'test_preparation_course',
                'math score':'math_score',
                'reading score':'reading_score',
                'writing score':'writing_score'})
stud_perform.head()
```

```python
stud_perform.groupby('gender').mean()

#as_index=1 or False -> flat table
mean_score=stud_perform.groupby(['gender','race/ethnicity'],as_index=1)\
    .aggregate({'math_score':'mean','reading_score':'mean'})\
    .rename(columns = {'math_score':'mean_math_score','reading_score':'mean_reading_score'})
mean_score
```

```python
mean_score.index
```

```python
mean_score.loc[[('female','group A'), ('female','group B')]]#if as_index=1
#mean_score.loc[stud_perform.gender == 'female',['race/ethnicity','mean_math_score']]#if as_index=0
```

```python
print(type(stud_perform.groupby(['gender','race/ethnicity']).math_score.nunique()))

stud_perform.groupby(['gender','race/ethnicity']).math_score.nunique()
```

```python
stud_perform.sort_values(['gender','math_score'],ascending=False)\
    .groupby(['gender']).head(5)
```

```python
stud_perform['total_score']=stud_perform.math_score+stud_perform.reading_score+stud_perform.writing_score
```

```python
stud_perform.head()
```

```python
#assign - allow to somehow change our data (for more complicated operations)
stud_perform=stud_perform.assign(total_score_log=np.log(stud_perform.total_score))
stud_perform.head()
```

```python
stud_perform.drop(['total_score_log'],axis=1)#axis=1 - index of columns 
```
